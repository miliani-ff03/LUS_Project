```python
import pandas as pd
import re
```

```python
# for duplicate preferences if selected video isnt video path with highest number flag

df_loaded = pd.read_csv('data_tables/uhw_duplicate_video_preferences.csv')

for index, row in df_loaded.iterrows():
    video_paths = row['all_videos'].strip("[]").replace("'", "").split(", ")
    max_num = -1
    preferred_video = None
    for vp in video_paths:
        # Search for any sequence of digits at the end of the filename before the extension
        match = re.search(r'(\d+)\.[^.]+$', vp)
        if match:
            num = int(match.group(1))
            if num > max_num:
                max_num = num
                preferred_video = vp
    
    if row['preferred'] == preferred_video:
        df_loaded.at[index, 'last_video_preferred'] = True
    else:
        df_loaded.at[index, 'last_video_preferred'] = False
# for video list identify video path that is last so ends with largest number
```

```python
# remove composite score column
df_loaded = df_loaded.drop(columns=['selection_method'])
```
